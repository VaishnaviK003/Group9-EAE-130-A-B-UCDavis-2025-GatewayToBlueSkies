```python
# Metabook pg. 98
# Using A3 for constants

import numpy as math
import matplotlib.pyplot as plt

w_empty, x_empty = 5833, 15.65935196
w_fuel, x_fuel = 994, 15.25
w_crew, x_crew = 180, 13
w_pay, x_pay = 2000, 18.62
x_full = ((w_empty * x_empty) + (w_fuel * x_fuel) +(w_crew * x_crew) +(w_pay * x_pay)) / (w_empty + w_fuel + w_crew + w_pay)
x_no_pay = ((w_empty * x_empty) + (w_fuel * x_fuel) +(w_crew * x_crew)) / (w_empty + w_fuel + w_crew)
x_no_fuel = ((w_empty * x_empty) +(w_crew * x_crew)) / (w_empty + w_crew)

print(w_empty + w_fuel + w_crew + w_pay, x_full/8)
print(w_empty + w_fuel + w_crew , x_no_pay/8)
print(w_empty + w_crew , x_no_fuel/8)
print(w_empty, x_empty/8)

weight = [w_empty + w_fuel + w_crew + w_pay,w_empty + w_fuel + w_crew,w_empty + w_crew, w_empty]
x = [x_full/8, x_no_pay/8, x_no_fuel/8, x_empty/8]
print(x_empty)
print(x_no_fuel)
print(x_no_pay)
print(x_full)
```

```python
plt.figure(figsize=(16,8))
plt.ylabel("Weight $(lbs)$", fontsize=40)
plt.xlabel("CG (Fraction of Mean Aerodynamic Chord)", fontsize=40)
plt.plot(x, weight, label='Center of Gravity Travel', linestyle='-', linewidth=8)
#plt.plot(426.923, 426.923 * (13 / 11.5), label='Intersection', marker="o")
plt.plot(x_full/8, w_empty + w_fuel + w_crew + w_pay, label='Empty Aircraft, Pilot, Fuel, Payload', marker="o", markersize=30)
plt.plot(x_no_pay/8, w_empty + w_fuel + w_crew, label='Empty Aircraft, Pilot, Fuel', marker="o", markersize=30)
plt.plot(x_no_fuel/8, w_empty + w_crew, label='Empty Aircraft, Pilot', marker="o", markersize=30)
plt.plot(x_empty/8, w_empty , label='Empty Aircraft', marker="o", markersize=30)

plt.ylim(5650, 9150)
plt.xlim(1.94,2.03)
plt.yticks(math.arange(5800, 9300, step=400), fontsize=25)
plt.xticks(math.arange(1.94, 2.035, step=0.01), fontsize=25)
plt.grid()
plt.legend(loc="best", fontsize=30)
plt.tight_layout()
plt.show()
```
