```python
# Metabook pg. 98
# Using A3 for constants

import numpy as math
import matplotlib.pyplot as plt

w_empty, x_empty = 5222, 15.15
w_fuel, x_fuel = 1350, 15.75
w_crew, x_crew = 180, 15.3
w_pay, x_pay = 2000, 5.8
x_full = ((w_empty * x_empty) + (w_fuel * x_fuel) +(w_crew * x_crew) +(w_pay * x_pay)) / (w_empty + w_fuel + w_crew + w_pay)
x_no_pay = ((w_empty * x_empty) + (w_fuel * x_fuel) +(w_crew * x_crew)) / (w_empty + w_fuel + w_crew)
x_no_fuel = ((w_empty * x_empty) +(w_crew * x_crew)) / (w_empty + w_crew)

print(w_empty + w_fuel + w_crew + w_pay, x_full/7)
print(w_empty + w_fuel + w_crew , x_no_pay/7)
print(w_empty + w_crew , x_no_fuel/7)
print(w_empty, x_empty/7)

weight = [w_empty + w_fuel + w_crew + w_pay,w_empty + w_fuel + w_crew,w_empty + w_crew, w_empty]
x = [x_full/7, x_no_pay/7, x_no_fuel/7, x_empty/7]
print(x_no_pay)
print(x_full)
```

```python
plt.figure(figsize=(20,16))
plt.ylabel("Weight $(lbs)$", fontsize=40)
plt.xlabel("Center of Gravity (Fraction of Mean Aerodynamic Chord)", fontsize=40)
plt.plot(x, weight, label='Center of Gravity Travel', linestyle='-', linewidth=8)
#plt.plot(426.923, 426.923 * (13 / 11.5), label='Intersection', marker="o")
plt.plot(x_full/7, w_empty + w_fuel + w_crew + w_pay, label='Empty Aircraft, Pilot, Fuel, Payload', marker="o", markersize=30)
plt.plot(x_no_pay/7, w_empty + w_fuel + w_crew, label='Empty Aircraft, Pilot, Fuel', marker="o", markersize=30)
plt.plot(x_no_fuel/7, w_empty + w_crew, label='Empty Aircraft, Pilot', marker="o", markersize=30)
plt.plot(x_empty/7, w_empty , label='Empty Aircraft', marker="o", markersize=30)

plt.ylim(5100, 8850)
plt.xlim(1.85,2.3)
plt.yticks(math.arange(5200, 9200, step=400), fontsize=30)
plt.xticks(math.arange(1.85, 2.325, step=0.05), fontsize=30)
plt.grid()
plt.legend(loc="best", fontsize=40)
plt.tight_layout()
plt.show()
```
